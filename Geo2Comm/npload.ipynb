```python
import os
import numpy as np
import scipy.io as sio
```

```python
os.chdir("/home/ub2204/桌面/g2c/SISO_simulate_v2")
```

```python
zero_gain_ue = np.load("zero_gain_ue.npy")
print(zero_gain_ue.shape)
print(zero_gain_ue)
print(len(zero_gain_ue))
```

```python
ue_pos_10k = np.load("ue_pos_10k.npy")
print(ue_pos_10k.shape)
print(ue_pos_10k)
print(len(ue_pos_10k))
print(ue_pos_10k[0])
```

```python
# set to 0 if the gain is 0

for i in range(len(zero_gain_ue)):
    bad_ue = zero_gain_ue[i]
    ue_pos_10k[bad_ue][2] = 0
```

```python
ue_pos_true = []
for i in range(len(ue_pos_10k)):
    if ue_pos_10k[i][2] != 0:
        ue_pos_true.append(ue_pos_10k[i])

ue_pos_true = np.array(ue_pos_true).astype(np.float64) 
```

```python
print(ue_pos_true.shape)
print(len(ue_pos_true))
```

```python
tx = np.array([149.5, -97.1, 26]).astype(np.float64) 
tx_list = []
tx_list.append(tx)
pos_dict = {
    'agent_num' : len(ue_pos_true),
    'agent' : ue_pos_true,
    'anchor_num' : 1,
    'anchor' : tx_list
}
    
sio.savemat(f'position1.mat', pos_dict)

np.save("ue_pos_true.npy", ue_pos_true)
```

```python
input_dir = "/home/ub2204/桌面/g2c/SISO_simulate_v2/data"
output_dir = "/home/ub2204/桌面/g2c/SISO_simulate_v2/data_mat"
```

```python
i=1
bad_ue_set = set(zero_gain_ue)
for file in os.listdir(input_dir):
    UE = int(file[:-4].split("_")[-1])
    if UE not in bad_ue_set :
        data = np.load(os.path.join(input_dir, file)).transpose()
        data_dict = {
                    # path index: from 1 to 297
                    'path_index': np.arange(1, 298).reshape(-1, 1),
                    'path_gain': data[0].reshape(-1, 1),
                    'path_phase': data[1].reshape(-1, 1),
                    'path_delay': data[2].reshape(-1, 1),
                    'path_AOA_hor':data[3].reshape(-1, 1),
                    'path_AOA_ver': data[4].reshape(-1, 1),
                    'path_AOD_hor': data[5].reshape(-1, 1),
                    'path_AOD_ver': data[6].reshape(-1, 1),
                    }

        mat_struct = {'sim': data_dict}
        sio.savemat(f'data_mat/data{i}.mat', mat_struct)    
        print(f"save data{i}.mat")
        i+=1
    
```
