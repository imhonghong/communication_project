```python
import os
import numpy as np
```

```python
x_min = -565.5
x_max = 563.6
y_min = -602
y_max = 600
```

```python
# grid size 128*128
grid_size = 128
x_step = (x_max - x_min) / grid_size
y_step = (y_max - y_min) / grid_size

grid_x = np.linspace(x_min, x_max, grid_size)
grid_y = np.linspace(y_min, y_max, grid_size)

# create grid
ue_pos_grid = []

for i in range(grid_size):
    for j in range(grid_size):
        x = grid_x[i]
        y = grid_y[j]
        ue_pos_grid.append([x, y, 1.5])

ue_pos_grid = np.array(ue_pos_grid)

# save grid
np.save('ue_pos_grid.npy', ue_pos_grid)

print(ue_pos_grid)
print(ue_pos_grid.shape)

# show grid
import matplotlib.pyplot as plt
plt.imshow(np.rot90(ue_pos_grid,k=1), interpolation='none')
plt.colorbar()
plt.show()
```
